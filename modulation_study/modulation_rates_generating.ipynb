The following generates all rates (1-10e) for all possible combinations of form factors, time of year, screening, that we have data for. Basically if you are trying to reproduce figures from our modulation paper and you don't have the rates, you can just run this next block of code. 

If you run this on a machine with a CUDA GPU, it will automatially use that, otherwise it defaults to using your CPU

from Modulation import generate_modulated_rates
overwrite = False
for doScreen in [True, False]: #remove the false here if you don't need unscreened rates
    for useVerne in [True,False]:
        for useQCDark in [True,False]: #using qcdark and qedark form factors
            for material in ['Si','Xe','Ar']: #Silicon, Xenon, Argon
                for FDMn in [0,2]: #heavy, light
                    if doScreen == False and useQCDark == True: #no point in doing qcdark rates unscreened, the main reason to do unscreened is to compare with old limits calculated with qedark that didn't include screening
                        continue
                    for summer in [False,True]: #note: summer data was only generated for DM masses 0.5-10MeV
                        if not useVerne and summer: #no damascus summer data was generated
                            continue
                        if summer and (material == 'Xe' or material == 'Ar'):
                            continue #since summer data only generated till 10 MeV, not very useful
                        generate_modulated_rates(material,FDMn,useQCDark=useQCDark,useVerne=useVerne,doScreen=doScreen,overwrite=overwrite,summer=summer)


```python
#generating for our sensei analysis
from Modulation import generate_modulated_rates
overwrite = False
for doScreen in [True, False]: #remove the false here if you don't need unscreened rates
    for useVerne in [True]:
        for useQCDark in [True,False]: #using qcdark and qedark form factors
            for material in ['Si']: #Silicon, Xenon, Argon
                for FDMn in [0,2]: #heavy, light
                    if doScreen == False and useQCDark == True: #no point in doing qcdark rates unscreened, the main reason to do unscreened is to compare with old limits calculated with qedark that didn't include screening
                        continue
                   
                    generate_modulated_rates(material,FDMn,useQCDark=useQCDark,useVerne=useVerne,doScreen=doScreen,overwrite=overwrite,summer=False,outdir='./modulation_analysis/')
```
